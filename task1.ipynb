```python
import numpy as np
import pandas as pd

#Loading dataset and converting into data frames
#data has seperated values byseperated by"\t"
df=pd.read_csv("C:\\Users\\Administrator\\Desktop\\projects\\data analyst internship\\task1\\marketing_campaign.csv",sep='\t');

#checking for null values 
#isnull() function identifies the rows with null values while sum() finds sum of all null values
print(df.isnull().sum());
```

```python
#droping duplicate values
df=df.drop_duplicates()
print(df)
#no duplicates are found
```

```python
# removing null values using dropna() and updating the dataframe df
df=df.dropna();
df.info()
```

```python
#we have 3 columns with object datatype
# analyzing column eductation 
distinct_values1= df["Education"].unique()
print(distinct_values1)
# First, convert to lowercase and strip whitespace
df['Education'] = df['Education'].str.lower().str.strip()

# Map the values to standardized ones
education_mapping = {
    'graduation': 'graduate',
    'phd': 'phd',
    'master': 'master',
    'basic': 'basic',
    '2n cycle': 'postgraduate'
}

df['Education'] = df['Education'].replace(education_mapping)
distinct_values2= df["Education"].unique()
print(distinct_values2)
```

```python
# analyzing column  Marital_Status
distinct_values1= df["Marital_Status"].unique()
print(distinct_values1)

# Convert to lowercase and remove extra spaces
df['Marital_Status'] = df['Marital_Status'].str.lower().str.strip()

# Create mapping dictionary
marital_mapping = {
    'married': 'married',
    'together': 'partnered',
    'single': 'single',
    'divorced': 'divorced',
    'widow': 'widowed',
    'alone': 'single',
    'absurd': 'other',
    'yolo': 'other'
}

# Apply the mapping
df['Marital_Status'] = df['Marital_Status'].replace(marital_mapping)
distinct_values2= df["Marital_Status"].unique()
print(distinct_values2)
```

```python
# Convert column Dt_customer object type to datetime type
df['Dt_Customer']

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y', errors='coerce')
print(df['Dt_Customer'])
```

```python
# Clean column names: lowercase, replace spaces with underscores
print(df.columns)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.columns)
#no changes in column names

#renaming the columns
df = df.rename(columns={
    'mntwines': 'mnt_wines',
    'mntfishproducts': 'mnt_fish_products',
    'numdealspurchases': 'num_deals_purchases',
    # Add more if needed
})
print(df.columns)
```

```python
#Check data types of all columns
print(df.dtypes)

#converting year into age 
from datetime import datetime

current_year = datetime.now().year
df['age'] = current_year - df['year_birth']
df['age'] = df['age'].astype(int)


#converting binary column to int not necessary as they are already in int datatype
binary_cols = ['acceptedcmp1', 'acceptedcmp2', 'acceptedcmp3', 'acceptedcmp4',
               'acceptedcmp5', 'response', 'complain']

for col in binary_cols:
    df[col] = df[col].astype(int)
    
#converting object into category to save memory
cat_cols = ['education', 'marital_status']
df[cat_cols] = df[cat_cols].astype('category')

print(df.dtypes)
```
